# LOADING THE DATA 

```python
import os
import cv2
import numpy as np
import math
import pandas as pd
import matplotlib.pyplot as plt

def slope(x1, y1, x2, y2): # Line slope given two points:
    s = (y2-y1)/(x2-x1)
    return 0 if math.isinf(s) else s

def find_angle(s1, s2): 
    return math.degrees(math.atan((s2-s1)/(1+(s2*s1))))%360

def load_data(data_dir, max_images_per_class=180):
    node_features_list = []
    label_vector = []
    d = 0

    for class_dir in os.scandir(data_dir):
        if not class_dir.is_dir():
            continue

        num_images_processed = 0
        class_label = int(os.path.basename(class_dir.path))
        for img_file in os.scandir(class_dir.path):
            if not img_file.name.endswith('.bmp'):
                continue
                
            num_images_processed += 1
            if num_images_processed >= max_images_per_class:
                break

            # Load the image
            img = cv2.imread(img_file.path)

            # Convert the image to grayscale
            gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

            # Threshold the image to create a binary image
            ret, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY_INV)

            # Find the contour of the Tamil alphabet
            contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_NONE)


            kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (19, 19))
            #morphological functions, dilation
            morph = cv2.morphologyEx(thresh, cv2.MORPH_DILATE, kernel)
            
            blank = np.zeros(thresh.shape[:2], dtype='uint8')
 
            cv2.drawContours(blank, contours, -1, (255, 0, 0), 1)
            
            # Discard contours with length less than 10
            if len(contours[0]) < 10:
                d += 1
                continue


            # Find the centroid (Ω) of the contour
            centroid_x = centroid_y = 0
            for contour in contours:
                moments = cv2.moments(contour)
                if moments["m00"] != 0:
                    centroid_x += int(moments["m10"] / moments["m00"])
                    centroid_y += int(moments["m01"] / moments["m00"])
                    
            centroid_x /= len(contours)
            centroid_y /= len(contours)
            centroid = (centroid_x, centroid_y)
            
            if num_images_processed == 1:
                plt.plot(centroid_x, centroid_y, color='red', marker='o', markersize=10)

            # Segment the contour into N different segments
            N = 20
            if num_images_processed == 1:
                print(len(contours))
#             if len(contours) > 2:
#                 continue
            segments = []
            for i in range(1):
                if len(contours[i]) == 0:
                    raise ValueError("Contour length should not be zero")
                segment_length = int(len(contours[i]) / N)
                if segment_length == 0:
                    segment_length = 1
                segments += [contours[i][j] for j in range(0, len(contours[i]), segment_length)]
            if len(segments) < N:
                d += 1
                continue
            seg_len = len(segments)
            segments = segments[:N]
            # Initialize arrays to store the extracted features
            lengths = []
            angles = []
            ratios = []
            # Loop through the segments
            for i in range(len(segments)):
                # Calculate the length (ρ) of the segment
                segment_length = math.dist(segments[i][0], centroid)
                lengths.append(segment_length)

                # Calculate the angle (θ) of the segment
                v1 = [(segments[i][0][0] - segments[(i+1)%N][0][0]), (segments[i][0][1] - segments[(i+1)%N][0][1])]
                v2 = [(segments[i][0][0] - centroid[0]), (segments[i][0][1] - centroid[1])]

                slope1 = slope(segments[i][0][0],segments[i][0][1],segments[(i+1)%N][0][0],segments[(i+1)%N][0][1])
                slope2 = slope(segments[i][0][0],segments[i][0][1],centroid[0],centroid[1])
                
                if num_images_processed == 1:
                    plt.plot([segments[i][0][0],centroid[0]],[segments[i][0][1],centroid[1]], color='white', linewidth=3)
                    plt.plot([segments[i][0][0],segments[(i+1)%N][0][0]],[segments[i][0][1],segments[(i+1)%N][0][1]], color='red', linewidth=3)
                
                ang_deg = find_angle(slope1, slope2)
                angle = 0
                if ang_deg-180>=0:
                    angle = 360 - ang_deg
                else: 
                    angle = ang_deg

                angles.append(angle)

                # Calculate the chord-to-arc-ratio (τ) of the segment
                chord_length = math.dist(segments[i][0], segments[(i+1)%N][0])
                eps = 1e-9
                ratio = chord_length / (segment_length + eps)
                ratios.append(ratio)
                

            if num_images_processed == 1:
                plt.imshow(blank)
                plt.show()

            feature_vector = np.array([lengths, angles, ratios]).flatten()
            if np.isnan(np.min(feature_vector)):
                if np.isnan(np.min(lengths)) or np.isnan(np.min(angles)) or np.isnan(np.min(ratios)):
                    d += 1
            if not np.isnan(np.min(feature_vector)):
                node_features_list.append(feature_vector)

                # Get the label for the image
                label = int(str(os.path.basename(class_dir.path)))
                label_vector.append(label-1)

    X = np.array(node_features_list)
    y = np.array(label_vector)
    print(d)
    return X, y 
```

# OPTIMIZED RANDOM FOREST

```python
import optuna
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

# Load the data
X, y = load_data(data_dir, 1000)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Define the objective function that takes a trial as input and returns a score
def objective(trial):
    # Define the search space for the hyperparameters
    n_estimators = trial.suggest_int('n_estimators', 50, 200)
    max_depth = trial.suggest_int('max_depth', 1, 50)
    min_samples_split = trial.suggest_int('min_samples_split', 2, 20)
    min_samples_leaf = trial.suggest_int('min_samples_leaf', 1, 10)
    
    # Create a Random Forest Classifier object with the hyperparameters suggested by Optuna
    rf = RandomForestClassifier(n_estimators=n_estimators, max_depth=max_depth, min_samples_split=min_samples_split, min_samples_leaf=min_samples_leaf, random_state=42)
    
    # Train the Random Forest Classifier on the training data
    rf.fit(X_train, y_train)
    
    # Test the Random Forest Classifier on the testing data and return the accuracy
    accuracy = rf.score(X_test, y_test)
    return accuracy

# Use Optuna to search for the optimal hyperparameters of the Random Forest Classifier
study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=100)

# Print the best hyperparameters found by Optuna
best_params = study.best_params
print('Best params:', best_params)

# Train the Random Forest Classifier on the training data with the optimal hyperparameters found by Optuna
rf = RandomForestClassifier(**best_params, random_state=42)
rf.fit(X_train, y_train)

# Test the Random Forest Classifier on the testing data and evaluate its performance
accuracy = rf.score(X_test, y_test)
print('Accuracy:', accuracy)
```

# OPTIMIZED XGB

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier

X, y = load_data(data_dir, 2160)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3
                                                    , random_state=42)

def objective(trial):
    train_x, valid_x, train_y, valid_y = train_test_split(X_train, y_train, test_size=0.2)

    param = {
        "verbosity": 0,
        "objective": "multclass:logistic",
        # use exact for small dataset.
        # "tree_method": "exact",
        # defines booster, gblinear for linear functions.
        "booster": trial.suggest_categorical("booster", ["gbtree", "gblinear", "dart"]),
        # L2 regularization weight.
        "lambda": trial.suggest_float("lambda", 1e-8, 1.0, log=True),
        # L1 regularization weight.
        "alpha": trial.suggest_float("alpha", 1e-8, 1.0, log=True),
        # sampling ratio for training data.
        "subsample": trial.suggest_float("subsample", 0.2, 1.0),
        # sampling according to each tree.
        "colsample_bytree": trial.suggest_float("colsample_bytree", 0.2, 1.0),
    }

    if param["booster"] in ["gbtree", "dart"]:
        # maximum depth of the tree, signifies complexity of the tree.
        param["max_depth"] = trial.suggest_int("max_depth", 3, 9, step=2)
        # minimum child weight, larger the term more conservative the tree.
        param["min_child_weight"] = trial.suggest_int("min_child_weight", 2, 10)
        param["eta"] = trial.suggest_float("eta", 1e-8, 1.0, log=True)
        # defines how selective algorithm is.
        param["gamma"] = trial.suggest_float("gamma", 1e-8, 1.0, log=True)
        param["grow_policy"] = trial.suggest_categorical("grow_policy", ["depthwise", "lossguide"])

    if param["booster"] == "dart":
        param["sample_type"] = trial.suggest_categorical("sample_type", ["uniform", "weighted"])
        param["normalize_type"] = trial.suggest_categorical("normalize_type", ["tree", "forest"])
        param["rate_drop"] = trial.suggest_float("rate_drop", 1e-8, 1.0, log=True)
        param["skip_drop"] = trial.suggest_float("skip_drop", 1e-8, 1.0, log=True)

    xgb = XGBClassifier(param)
    xgb.fit(train_x, train_y)
    accuracy = xgb.score(valid_x, valid_y)
    return accuracy

# 3. Create a study object and optimize the objective function.
study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=100)
```

```python
import lightgbm as lgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_curve, auc
from sklearn.preprocessing import label_binarize
import matplotlib.pyplot as plt

lr = 1e-4
X, y = load_data(data_dir, 2160)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3
                                                    , random_state=42)

clf = lgb.LGBMClassifier(boosting_type='goss', num_leaves=25, max_depth=-1, learning_rate=0.09, n_estimators=100, random_state=42)

clf.fit(X_train, y_train)

accuracy = clf.score(X_test, y_test)

y_prob = clf.predict_proba(X_test)

y_test_bin = label_binarize(y_test, classes=clf.classes_)

fpr = dict()
tpr = dict()
roc_auc = dict()
for i in range(len(clf.classes_)):
    fpr[i], tpr[i], _ = roc_curve(y_test_bin[:, i], y_prob[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

plt.figure()
lw = 2
for i in range(len(clf.classes_)):
    plt.plot(fpr[i], tpr[i], lw=lw, label='ROC curve of class {0} (AUC = {1:0.2f})'.format(clf.classes_[i], roc_auc[i]))

plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()
```
